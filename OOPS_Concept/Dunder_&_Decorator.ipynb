# Dunder method
start with (__)
__init__

__str__
__len__
__add__

They are also called magic methods or special methods.
Dunder methods = “special hooks” that Python automatically calls

```python
class A:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return self.x + other.x

a1 = A(5)
a2 = A(10)
print(a1 + a2)  # 15
```

### *args and **kwargs
*args is used to pass multiple positional arguments (non-keyword arguments).
**kwargs is used to pass multiple keyword arguments (key-value pairs).

```python
def add(*args):
        sum=0
        for i in args:
            sum=sum+i
        print(sum)
add(1,2,2,6,9)
```

```python
def information(**kwargs):
        print("your info : ")
        for i in kwargs:
            print(i,kwargs[i])
information(name="Sanya",age=34)
```

### Decorators
A decorator is just a function that takes another function and extends its behavior.

```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper
@my_decorator
def hello():
    print("Hello")

hello()
```
