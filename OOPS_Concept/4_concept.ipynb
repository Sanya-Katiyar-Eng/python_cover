##  Polymorphism
same method behave different
Types of Polymorphism in Python

Python has 4 main forms:

Method Overriding
Method Overloading (simulated)
Operator Overloading
Duck Typing

```python
#Method Overriding 
# child class changes the behavior of parent class method.
#Why it is polymorphism?

#Same method name sound()
#👉 Different behavior depending on object
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")
a = Animal()
a.sound()

d = Dog()
d.sound()
```

### Method Overloading
Same function name with different parameters.
Python does NOT support it directly.❌

### Operator Overloading
Same operator behaves differently depending on objects.


```python
print(10 + 20)          # addition
print("A" + "B")        # string concat
```

### Duck Typing
“If it looks like a duck and quacks like a duck, it is a duck”

👉 Python doesn’t care about type
👉 Only cares about behavior (methods)

```python
class Car:
    def move(self):
        print("Car is moving")

class Plane:
    def move(self):
        print("Plane is flying")
def start_vehicle(vehicle):
    vehicle.move()
start_vehicle(Car())
start_vehicle(Plane())
```

3. Private(use under the class)

```python
class Animal:
    def __init__(self):
        self.name="Sanya"
        self._age=23
        self.__salary=23
    def show(self):
        print("Name :",self.name)
        print("Age : ",self._age)
        print("Salary : ",self.__salary)
class s(Animal):
     def s1(self):
         print("name")

   
a=s()
a.show()
#print("public argument :",a.name)
```

```python
class Student:
    def __init__(self):
        self.__marks = 90

    @property
    def marks(self):
        return self.__marks

    @marks.setter
    def marks(self, value):
        if value > 0:
            self.__marks = value
s = Student()
s.marks = 100
print(s.marks)
```
